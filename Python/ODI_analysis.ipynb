```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

path="/content/drive/MyDrive/ODI data.csv"
df=pd.read_csv(path)
print(df.head())
```

```python
df.info()
```

```python
df.describe()
```

```python
da=df.head(10)
d = pd.DataFrame(da, columns=["Runs", "Player"])
d
```

```python
d = pd.DataFrame(da, columns=["100", "Player"])
d
sorted_data=d.sort_values(by="100",ascending=False)
sorted_data
```

```python
sns.barplot(x="Player",y="Runs",data=da)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=da.groupby('Player')['100'].sum().reset_index().sort_values('100', ascending=False),
            x='100', y='Player')
plt.title('Most ODI Centuries')
plt.xlabel('100')
plt.ylabel('Player')
plt.tight_layout()
plt.show()
```
